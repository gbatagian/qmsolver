```python
from qmsolver.tise import FDSolver
from qmsolver.potentials import DoubleSquareWellPotential
```

```python
dw_solver = FDSolver(steps=2_000, x_min=-5, x_max=5, n_lowest=12)
dw_potential = DoubleSquareWellPotential(
    x_grid=dw_solver.x_grid,
    well_1_center=-2,
    well_1_depth=25,
    well_1_width=2,
    well_2_center=2,
    well_2_depth=25,
    well_2_width=2,
)
dw_solver.potential_generator = dw_potential
dw_solver.solve()
dw_solver.output()
dw_solver.plot()
```
