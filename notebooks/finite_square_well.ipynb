```python
from qmsolver.tise import FDSolver
from qmsolver.potentials import FiniteSquareWellPotential
```

```python
fsw_solver = FDSolver(steps=2_000, x_min=-5, x_max=5, n_lowest=6)
fsw_potential = FiniteSquareWellPotential(
    x_grid=fsw_solver.x_grid, well_depth=25, well_width=2
)
fsw_solver.potential_generator = fsw_potential
fsw_solver.solve()
fsw_solver.output()
fsw_solver.plot()
```
