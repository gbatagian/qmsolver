```python
from qmsolver.tise import FDSolver
from qmsolver.potentials import HarmonicOscillatorPotential
```

```python
ho_solver = FDSolver(steps=2_000, x_min=-5, x_max=5, n_lowest=5)

# The `grid_active_range` parameter controls the width of the harmonic well, the potential is non zero only within the specified range
# around the center of the grid.
ho_potential = HarmonicOscillatorPotential(
    x_grid=ho_solver.x_grid,
    spring_constant=1.0,
    grid_active_range=0.42,  # potential well spans 42% of grid length, centered
)

ho_solver.potential_generator = ho_potential
ho_solver.solve()
ho_solver.output()
ho_solver.plot()
```

```python
ho_solver = FDSolver(steps=2_000, x_min=-15, x_max=-5, n_lowest=5)

# If not specified, default value of `grid_active_range` is 0.5. This ensures that the potential extends sufficiently beyond the
# range of the well, so that bound states naturally decay to zero at the grid edges and the FDSolver solutions are meaningful.
ho_potential = HarmonicOscillatorPotential(
    x_grid=ho_solver.x_grid, spring_constant=1.0
)
ho_solver.potential_generator = ho_potential
ho_solver.solve()
ho_solver.output()
ho_solver.plot()
```
