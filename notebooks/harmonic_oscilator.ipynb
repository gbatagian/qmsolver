```python
from qmsolver.tise import FDSolver
from qmsolver.potentials import HarmonicOscillatorPotential
```

```python
ho_solver = FDSolver(steps=2_000, x_min=-5, x_max=5, n_lowest=12)
ho_potential = HarmonicOscillatorPotential(
    x_grid=ho_solver.x_grid, spring_constant=1.0
)
ho_solver.potential_generator = ho_potential
ho_solver.solve()
ho_solver.output()
ho_solver.plot()
```

```python
ho_solver = FDSolver(steps=2_000, x_min=-5, x_max=5, n_lowest=20)
ho_potential = HarmonicOscillatorPotential.from_frequency(
    x_grid=ho_solver.x_grid, frequency=1.5
)
ho_solver.potential_generator = ho_potential
ho_solver.solve()
ho_solver.output()
ho_solver.plot()
```
