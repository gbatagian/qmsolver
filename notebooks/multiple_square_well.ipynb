```python
from qmsolver.tise import FDSolver
from qmsolver.potentials import MultipleSquareWellPotential
```

```python
mw_solver = FDSolver(steps=4_000, x_min=-16, x_max=16, n_lowest=22)
mw_potential = MultipleSquareWellPotential(
    x_grid=mw_solver.x_grid,
    well_depth=15,
    well_width=1.0,
    separation=2.5,
    num_wells=10,
)
mw_solver.potential_generator = mw_potential
mw_solver.solve()
mw_solver.output()
mw_solver.plot()
```
