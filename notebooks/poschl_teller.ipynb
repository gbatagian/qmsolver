```python
from qmsolver.tise import FDSolver
from qmsolver.potentials import PoschlTellerPotential
```

```python
pt_solver = FDSolver(steps=2000, x_min=-8, x_max=8, n_lowest=6)
pt_potential = PoschlTellerPotential(x_grid=pt_solver.x_grid, lambda_=5)
pt_solver.potential_generator = pt_potential
pt_solver.solve()
pt_solver.output()
pt_solver.plot()
```
